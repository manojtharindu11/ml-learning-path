```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
df0 = df[df.target == 0]
df1 = df[df.target == 1]
df2 = df[df.target == 2]
```

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color='blue', label='setosa', marker='+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color='green', label='versicolor', marker='.')
```

```python
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color='blue', label='setosa', marker='+')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], color='green', label='versicolor', marker='.')
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['target', 'flower_name'], axis=1)
X.head()
```

```python
Y = df['target']
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```
