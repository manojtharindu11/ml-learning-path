```python
df = pd.read_csv("csv/test_scores.csv")
df
```

```python
maths = df["math"]
cs = df["cs"]
```

```python
import math
```

```python
def gradient_descent(x, y):
    # Initialize parameters for the line: slope (m_current) and intercept (y_current)
    m_current = y_current = 0
    iteration = 100000
    n = len(x)
    learning_rate = 0.0001
    previous_cost = None  # To store the cost from the previous iteration for comparison
    
    for i in range(iteration):
        # Predict y values using current parameters
        y_predicted = m_current * x + y_current
        # Calculate mean squared error cost
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        
        # If not the first iteration, compare current cost with previous cost
        if previous_cost is not None:
            print(f"Cost difference: {cost - previous_cost}")  # Show how much cost changed
            if previous_cost < cost:
                # If cost increases, stop (gradient descent should always decrease cost)
                break
            isClose = math.isclose(cost, previous_cost, abs_tol=1e-5)  # Check for convergence
            if isClose:
                print(f"Converged at iteration {i+1} with cost {cost}")
                break
        previous_cost = cost  # Update previous_cost for next iteration
        
        # Calculate gradients for m and y (partial derivatives)
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        
        # Update parameters using gradients and learning rate
        m_current -= learning_rate * md
        y_current -= learning_rate * bd
        print(f"Iteration {i+1}: m = {m_current}, b = {y_current}, cost = {cost}")
    
gradient_descent(maths, cs)
```
