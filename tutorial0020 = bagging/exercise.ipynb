```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df4.Sex = le.fit_transform(df4.Sex)
df4.ChestPainType = le.fit_transform(df4.ChestPainType)
df4.RestingECG = le.fit_transform(df4.RestingECG)
df4.ExerciseAngina = le.fit_transform(df4.ExerciseAngina)
df4.ST_Slope = le.fit_transform(df4.ST_Slope)
df4.head()
```

```python
X = df4.drop("HeartDisease",axis="columns")
Y = df4.HeartDisease
X.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Column indices: Sex=1, ChestPainType=2, RestingECG=6, ExerciseAngina=8, ST_Slope=10
ct = ColumnTransformer([
    ("categorical", OneHotEncoder(drop='first'), [1, 2, 6, 8, 10])
], remainder='passthrough')
X_transformed = ct.fit_transform(X)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_transformed)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=30)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score

scores = cross_val_score(SVC(), X, Y, cv=5)
scores
```

```python
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
    estimator=SVC(),
    n_estimators=100,
    max_samples=0.8,
    oob_score=True,
    random_state=0
)

bag_model.fit(X_train, Y_train)
```

```python
bag_model.score(X_test, Y_test)
```
